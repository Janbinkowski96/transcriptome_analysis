```python
import os
from math import log
from IPython.display import display

import pandas as pd
from utils.utils import load_from_dir, save_data
```

```python
def select_fc(df: pd.DataFrame, fc_column: str) -> pd.DataFrame:
    selected_df = df[df[fc_column].abs() > 2]
    display(selected_df.shape)
    return selected_df

def select_pval(df: pd.DataFrame, pval_column: str, threshold: float) -> pd.DataFrame:
    selected_df = df[df[pval_column] <= threshold]
    display(selected_df.shape)
    return selected_df
```

```python
data = load_from_dir("../data/raw/", suffix="xlsx")
```

```python
rxl = data.get("RXL")[["GO", "RXLbw/RXL.fc", "RXLbw/RXL.raw.pval", "N_RXL1", "N_RXL2", "N_RXLbw1", "N_RXLbw2"]]
display(rxl.head())

ds2 = data.get("DS2")[["GO", "DS2bw/DS2.fc", "DS2bw/DS2.raw.pval", "N_DS2_1", "N_DS2_2", "N_DS2bw1", "N_DS2bw2", "N_DS2bw3"]]
display(ds2.head())


m12 = data.get("M12")[["GO", "M12bw/M12wn.fc", "M12bw/M12wn.raw.pval", "N_M12bw1", "N_M12bw2", "N_M12bw3", "N_M12wn2", "N_M12wn3"]]
display(m12.head())
```

```python
save_data([rxl, ds2, m12], names=["RXL.csv", "DS2.csv", "M12.csv"], path="../data/interim")
```

```python
# RXL
rxl_selected_fc = select_fc(rxl, "RXLbw/RXL.fc")

rxl_selected_fc_05 = select_pval(rxl_selected_fc, "RXLbw/RXL.raw.pval", threshold=0.05)
rxl_selected_fc_005 = select_pval(rxl_selected_fc, "RXLbw/RXL.raw.pval", threshold=0.005)
rxl_selected_fc_001 = select_pval(rxl_selected_fc, "RXLbw/RXL.raw.pval", threshold=0.001)

save_data([rxl_selected_fc_05, rxl_selected_fc_005, rxl_selected_fc_001], names=["RXL_05.csv", "RXL_005.csv", "RXL_001.csv"], path="../data/prepared/")
```

```python
# DS2
ds2_selected_fc = select_fc(ds2, "DS2bw/DS2.fc")

ds2_selected_fc_05 = select_pval(ds2_selected_fc, "DS2bw/DS2.raw.pval", threshold=0.05)
ds2_selected_fc_005 = select_pval(ds2_selected_fc, "DS2bw/DS2.raw.pval", threshold=0.005)
ds2_selected_fc_001 = select_pval(ds2_selected_fc, "DS2bw/DS2.raw.pval", threshold=0.001)

save_data([ds2_selected_fc_05, ds2_selected_fc_005, ds2_selected_fc_001], names=["DS2_05.csv", "DS2_005.csv", "DS2_001.csv"], path="../data/prepared/")
```

```python
# M12
m12_selected_fc = select_fc(m12, "M12bw/M12wn.fc")

m12_selected_fc_05 = select_pval(m12_selected_fc, "M12bw/M12wn.raw.pval", threshold=0.05)
m12_selected_fc_005 = select_pval(m12_selected_fc, "M12bw/M12wn.raw.pval", threshold=0.005)
m12_selected_fc_001 = select_pval(m12_selected_fc, "M12bw/M12wn.raw.pval", threshold=0.001)

save_data([m12_selected_fc_05, m12_selected_fc_005, m12_selected_fc_001], names=["M12_05.csv", "M12_005.csv", "M12_001.csv"], path="../data/prepared/")
```
